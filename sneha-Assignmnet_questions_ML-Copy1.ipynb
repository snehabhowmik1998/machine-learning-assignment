### What is the distribution  of preferred foot?

```python
preffered_foot=df2["Preferred Foot"].value_counts()
preffered_foot
```

```python
plt.figure(figsize=(7,6))
preffered_foot.plot(kind="bar", color="pink")
plt.xlabel("Distribution of preferred foot")
plt.ylabel("Frequency")
plt.show()
```

### How can we visualize the age distribution of players?

```python
#graph 1
plt.hist(df2["Age"])
plt.xlabel("Age")
plt.ylabel("number of players")
plt.title("Age Distribution of Players")
plt.show()
```

```python
#grsaph 2
sns.countplot(x=df2["Age"],data=df2)
plt.xlabel("Age")
plt.ylabel("number of players")
plt.title("Age Distribution of Players")
plt.show()
```

### What is the international reputation of left and right foot?

```python
sns.countplot(x=df2['International Reputation'], hue=df2['Preferred Foot'],  data=df2)
plt.title('International Reputation of Left and Right Footed Players ', fontsize = 12)
plt.show()
```

```python
df2.groupby('Preferred Foot')['International Reputation'].value_counts()
```

### What is the different work rate of the Players?

```python
work_rate_of_players=df2['Work Rate'].value_counts()
work_rate_of_players
```

### Top 10 countries with most number of players

```python
top_countries_name=df2["Nationality"].value_counts().head(10) # assigning top 10 values by using head() function
top_countries_name
```

```python
fig, ax = plt.subplots(figsize=(12,8))
x = top_countries_name.values# x axis
y = top_countries_name.index # y axis
ax.barh(y, x, align='center', color='yellow')
ax.set_xlabel('Number of Players')
ax.set_ylabel('Name of Countries')
ax.set_title('Top 10 Countries with most number of players')
plt.show()
```

### What is the work rate of players segregated by preferred foot?

```python
plt.figure(figsize=(10,7))# defininig figure size
sns.countplot(x=df2["Work Rate"] , hue=df2["Preferred Foot"] , data = df2)
plt.title("work rate of players segregated by preferred foot")
plt.show()
```

```python
df2.groupby("Preferred Foot")["Work Rate"].value_counts() #using groupby function
```

### Number of players at different positions? 



```python
plt.figure(figsize=(11,6))
sns.countplot(x=df2['Position'], data= df2, palette="Set3")
plt.title(label = 'Count of Players with Positions')
plt.xlabel(xlabel = 'Positions in Football')
plt.ylabel(ylabel = 'Number of Players')
plt.show()
```

```python
df2['Position'].value_counts()
```
