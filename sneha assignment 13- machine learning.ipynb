```python
df2["Plas"]=df2["Plas"].replace(0,df2["Plas"].mean())
df2["Pres"]=df2["Pres"].replace(0,df2["Pres"].mean())
df2["skin"]=df2["skin"].replace(0,df2["skin"].mean())
df2["test"]=df2["test"].replace(0,df2["test"].mean())
df2["mass"]=df2["mass"].replace(0,df2["mass"].mean())
df2["pedi"]=df2["pedi"].replace(0,df2["pedi"].mean())
# chenging 0 values to their respective mean values of the columns
```

```python
x=df2.iloc[:,0:8] # taking all independent variables
y=df2.iloc[:,-1] # taking dependent variables
```

```python
from sklearn.model_selection import train_test_split # importing train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=12)
```

```python
from sklearn.linear_model import LogisticRegression # importing logisticregression
from sklearn.metrics import accuracy_score # importing accuraccy_score 
```

```python
clf= LogisticRegression(random_state=0, solver='newton-cg', verbose=1)
```

```python
clf.fit(x_train,y_train)
```

```python
pred1=clf.predict(x_train)
```

```python
score1=accuracy_score(y_train,pred1)
```

```python
score1
print("score of y_train and pred1 is ",score1*100)
```

```python
pred2=clf.predict(x_test)
```

```python
score2=accuracy_score(y_test, pred2)
```

```python
score2
print("score of y_test and pred2 is ",score2*100)
```
