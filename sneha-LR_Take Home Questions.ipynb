### 5. Build classification model and present it's classification report ?

```python
df.head()
```

```python
df2 = df.drop('Class',axis = 1)
df_with_dummies = pd.get_dummies(df2, drop_first=True)
df_with_dummies.head()
```

```python
Features = df_with_dummies.drop(['numeric_class'],axis = 1)
Target = df_with_dummies['numeric_class']
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(Features)
```

```python
from sklearn.model_selection import train_test_split

X = scaler.fit_transform(Features)
X_train, X_test, y_train, y_test = train_test_split(X, Target, test_size=0.3, random_state=45)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
pred
```

```python
score1=lr.score(X_train,y_train)
score1
```

```python
score2=lr.score(X_test,y_test)
score2
```

```python
from sklearn.metrics import accuracy_score
sc=accuracy_score(y_test, pred)
sc
```

```python
from sklearn.metrics import classification_report 
cr=classification_report(y_test, pred)
print(cr)
```
